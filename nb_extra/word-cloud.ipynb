# Word Clouds

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

%matplotlib inline
```

```python
with open('../data/stopwords.txt', 'r', encoding='utf8') as f:
    stopwords = f.read().splitlines()
```

```python
df = pd.read_csv('../data/tokens-for-word-cloud.csv')
```

```python
# add new stopwords

new_stopwords = ['http', 'amp', 'the']
stopwords = [*stopwords, *new_stopwords]
```

```python
wc = WordCloud(width=3000, 
               height=2000,
               random_state=42,
               background_color='white', 
               colormap='magma', 
               collocations=False, 
               stopwords=stopwords,
               max_words=150)
wc.generate(df.loc[:, 'tokens'].str.cat(sep=' '))
plt.figure(figsize=(20, 20))
plt.imshow(wc, interpolation="bilinear")
plt.axis('off');
# plt.savefig('../img/Wordcloud-gesamt', dpi=300, bbox_inches='tight');
```
